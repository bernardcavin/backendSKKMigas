```python
import geopandas as gpd
from geoalchemy2 import Geometry
from sqlalchemy import create_engine
import json
import os
```

```python
def clean_string_replace(s, chars_to_remove):
    for char in chars_to_remove:
        s = s.replace(char, '') 
    return s

def save_attribute_based(gdf, attribute, folder):
    os.makedirs('data/'+folder, exist_ok=True)
    for i in range(len(gdf)):
        row_name = gdf.iloc[i].get(attribute)
        if row_name:
            row_name = clean_string_replace(row_name, chars_to_remove=['"', '\\', '/', ':', '*', '?', '<', '>', '|', ' ', '\n']) if type(row_name) == str else row_name
            row_json = gdf.iloc[[i]].to_json()
            with open(f'data/{folder}/{row_name}.json', 'w+') as f:
                f.write(row_json)
        else:
            pass
```
